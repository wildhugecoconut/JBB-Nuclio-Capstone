```python
def merge_and_add_region(df_base, df_merge, region_name):
    # Realizar el merge de los DataFrames
    df_merged = pd.merge(df_base, df_merge, on="Fecha", how="left")

    # Agregar la columna de región con el nombre especificado
    df_merged["Region"] = region_name

    return df_merged
```

```python
def T_en_df(df,country_name):
  df=df.T
  columnas_a_eliminar = [7, 8, 9]
  df.drop(df.columns[columnas_a_eliminar], axis=1, inplace=True)
  df.reset_index(inplace=True)
  df.columns = df.iloc[0]
  df = df[1:]
  df.fillna(country_name, inplace=True)
  df.rename(columns={df.columns[1]: "Region"}, inplace=True)
  df.set_index(df.columns[0], inplace=True)
  df["Regionn"] = df["Region"]
  return df
```

```python
def secont_df(df):
  df_mujeres = df.iloc[:, :4]
  df_hombres = df.iloc[:, 4:]

  df_hombres["gender"] = 0
  df_hombres["Region"] = df_hombres["Regionn"]
  del df_hombres["Regionn"]

  df_mujeres["gender"] = 1

  df_resultado = pd.concat([df_hombres, df_mujeres])
  df_resultado.drop(df_resultado.columns[0], axis=1, inplace=True)
  df_resultado.drop(df_resultado.columns[4], axis=1, inplace=True)
  df_resultado.reset_index(drop=False, inplace=True)

  df_resultado.rename(columns={df_resultado.columns[1]: 'Tasa de Paro en personas menores de 25 años'}, inplace=True)
  df_resultado.rename(columns={df_resultado.columns[2]: 'Tasa de Paro en personas de 25 y más años'}, inplace=True)

  return df_resultado
```
